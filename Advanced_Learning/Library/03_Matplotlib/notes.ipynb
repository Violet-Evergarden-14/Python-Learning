## Multi-figures
\
show in one figure

```python
x = [0, 1, 2, 3, 4]
y1 = [1, 2, 3, 4, 5]
y2 = [-i for i in y1]
y3 = [0 for i in y1]
```

```python
Fig1 = plt.figure()
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
```

```python
Fig2 = plt.figure()
axe2 = plt.axes()         # the two lines can be replaced with "Fig2, axe2 = plt.subplots()"
axe2.plot(x, y1)
axe2.plot(x, y2)
axe2.plot(x, y3)
```

show in different figures -- subfigrue

```python
x = [0, 1, 2, 3, 4]
y1 = [1, 2, 3, 4, 5]
y2 = [-i for i in y1]
y3 = [0 for i in y1]
```

```python
Fig1 = plt.figure()
plt.subplot(3, 1, 1), plt.plot(x, y1)
plt.subplot(3, 1, 2), plt.plot(x, y2)
plt.subplot(3, 1, 3), plt.plot(x, y3)
```

```python
Fig2, ax = plt.subplots(3)
ax[0].plot(x, y1)
ax[1].plot(x, y2)
ax[2].plot(x, y3)
```

## two-dimention figures

choose color, style, width, mark

```python
x = [0, 1, 2, 3, 4]
y = [1, 2, 3, 4, 5]
Fig = plt.figure()
plt.plot(x, y, color = '#47ADC7', linestyle = '--', linewidth = 2, marker = 'o', markersize = 10)
# if we'd like to plot dots, we can make the linestyle as '', while the marker is 'o'
```

## grid figure

```python
import numpy as np
x = np.linspace(0, 10, 1000)
f = np.sin(x) * np.cos(x).reshape(-1, 1)
```

```python
Fig1 = plt.figure()
plt.imshow(f)
plt.colorbar()
```

```python
Fig2, ax2 = plt.subplots()
ax2.imshow(f)
# it has no color bar
```

## statistics figure

```python
import numpy as np
data = np.random.randn(10000)
```

```python
Fig1 = plt.figure()
plt.hist(data)
```

```python
# the number of the bars -- bins
Fig2 = plt.figure()
plt.hist(data, bins = 30)
```

```python
# tarnsparency -- alpha
Fig3 = plt.figure()
plt.hist(data, alpha = 0.5)
```

```python
# type -- histtype
Fig4 = plt.figure()
plt.hist(data, histtype = 'stepfilled')
```

```python
# color -- color, edgecolor
Fig4 = plt.figure()
plt.hist(data, color = '#A074C4', edgecolor = '#FFFFFF')
```

## window attribute
### axis

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 8, 27, 64, 125]
```

lim

```python
Fig1 = plt.figure()
plt.plot(x, y)
plt.xlim(0, 5)
plt.ylim(0, 125)
```

```python
Fig2, axe2 = plt.subplots()
axe2.plot(x, y)
axe2.set_xlim(0, 5)
axe2.set_ylim(0, 125)
```

axis

```python
Fig1 = plt.figure()
plt.plot(x, y)
plt.axis([0, 5, 0, 125])     # axis('equal') can make x-axis and y-axis have the same length
```

```python
Fig2, axe2 = plt.subplots()
axe2.plot(x, y)
axe2.axis([0, 5, 0, 125])
```

### names

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 8, 27, 64, 125]
```

```python
Fig1 = plt.figure()
plt.plot(x, y)
plt.title('title')
plt.xlabel('xlable')
plt.ylabel('ylable')
```

```python
Fig2, axe2 = plt.subplots()
axe2.plot(x, y)
axe2.set_title('title')
axe2.set_xlabel('xlable')
axe2.set_ylabel('ylable')
```

### figure sample

```python
x = [1, 2, 3, 4, 5]
y1 = [i for i in x]
y2 = [0 for i in x]
y3 = [-i for i in x]
```

```python
Fig = plt.figure()
plt.plot(x, y1, label = 'y = x')
plt.plot(x, y2, label = 'y = 0')
plt.plot(x, y3, label = 'y = -x')
plt.legend()     # can also be .legend(['y = x', 'y = 0', 'y = -x'])
```

three key words:\
loc: choose the position of the label\
frameon: choose the frame\
ncol: the number of the columns of the label

### add grid

also has key words: color, linestyle

```python
x = [1, 2, 3, 4, 5]
y1 = [i for i in x]
y2 = [0 for i in x]
y3 = [-i for i in x]

Fig = plt.figure()
plt.plot(x, y1, label = 'y = x')
plt.plot(x, y2, label = 'y = 0')
plt.plot(x, y3, label = 'y = -x')
plt.legend()

plt.grid()
```
