### 2 ways to draw a picture
1st: matlab method

```python
Fig1 = plt.figure()
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 8, 27, 64, 125]
plt.plot(x, y)
```

2nd: object-oriented method

```python
Fig2 = plt.figure()
ax2 = plt.axes()
ax2.plot(x, y)
```
